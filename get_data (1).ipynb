```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
scripts = [
    ("https://imsdb.com/scripts/Fault-in-Our-Stars,-The.html", "fault_in_our_stars.txt"),
    ("https://imsdb.com/scripts/Perks-of-Being-a-Wallflower,-The.html", "perks_of_being_a_wallflower.txt"),
    ("https://imsdb.com/scripts/Pride-and-Prejudice.html", "pride_and_prejudice.txt"),
    ("https://imsdb.com/scripts/Titanic.html", "titanic.txt"),
    ("https://imsdb.com/scripts/10-Things-I-Hate-About-You.html", "10_things_i_hate_about_you.txt"),
    ("https://imsdb.com/scripts/Avatar.html", "avatar.txt"),
    ("https://imsdb.com/scripts/Fifth-Element,-The.html", "fifth_element.txt"),
    ("https://imsdb.com/scripts/Interstellar.html", "interstellar.txt"),
    ("https://imsdb.com/scripts/Tenet.html", "tenet.txt"),
    ("https://imsdb.com/scripts/Martian,-The.html", "martian.txt")
]
```

```python
for url, filename in scripts:
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        script_text = soup.find('pre').get_text()
        
        with open(filename, 'w', encoding='utf-8') as f:
            f.write(script_text)
        
        print(f"{filename} downloaded!")
        time.sleep(2)  
    except:
        print(f"{filename} not downloaded!")
```

```python
def split_script_into_blocks(script_text):
    
    script_text = re.sub(r'\n\s*\n\s*\n+', '\n\n', script_text)
    
    blocks = script_text.split('\n\n')
    
    valid_blocks = []
    for block in blocks:
        block = block.strip()
        if len(block) > 50:  
            valid_blocks.append(block)
    
    return valid_blocks
```

```python
categories = ['romantic', 'scifi']
```

```python
import re
```

```python
for category in categories:
    category_path = f'corpus/{category}'
    blocks_path = f'corpus/{category}_blocks'
    os.makedirs(blocks_path, exist_ok=True)
    
    total_blocks = 0
    
    for filename in os.listdir(category_path):
        if filename.endswith('.txt'):
            filepath = os.path.join(category_path, filename)
            
            with open(filepath, 'r', encoding='utf-8', errors='ignore') as f:
                script = f.read()
            
            blocks = split_script_into_blocks(script)
            
           
            movie_name = filename.replace('.txt', '')
            for idx, block in enumerate(blocks, 1):
                block_filename = f'{blocks_path}/{movie_name}_block_{idx:03d}.txt'
                with open(block_filename, 'w', encoding='utf-8') as f:
                    f.write(block)
            
            total_blocks += len(blocks)
            print(f"{filename}: {len(blocks)} blok")
    
    print(f"\n{category.upper()} Total: {total_blocks} documents\n")
    print("="*50)
```
